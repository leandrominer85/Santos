```python
# Matches, rounds List and results (santos goals - other team goals)
url = "https://sofascore.p.rapidapi.com/teams/get-last-matches"

querystring = {"teamId":"1968","pageIndex":"0"}

headers = {
    'x-rapidapi-host': "sofascore.p.rapidapi.com",
    'x-rapidapi-key': "d7f888003emshbc18d63df61e366p1d4c9ajsn104687195dea"
    }

response = requests.request("GET", url, headers=headers, params=querystring)

matches = []
rounds =[]
home= []
results = []
n= 1
for i in range(len(response.json()["events"])):
  if response.json()['events'][i]['tournament']['uniqueTournament']['id'] == 325:
    matches.append(response.json()['events'][i]['id'])
    rounds.append(f"Round {n}")
    n +=1
    if response.json()['events'][i]['homeTeam']['id'] == 1968:
      home.append('home')
      results.append(response.json()['events'][i]['homeScore']['current']  - response.json()['events'][i]['awayScore']['current'])
    else:
      home.append("away")
      results.append(response.json()['events'][i]['awayScore']['current']  - response.json()['events'][i]['homeScore']['current'])

    
```

```python
url = "https://sofascore.p.rapidapi.com/matches/get-statistics"



headers = {
    'x-rapidapi-host': "sofascore.p.rapidapi.com",
    'x-rapidapi-key': "d7f888003emshbc18d63df61e366p1d4c9ajsn104687195dea"
    }

names=[]
stats=[]
for idx , match in enumerate(matches):
  querystring = {"matchId":f"{match}"}
  response = requests.request("GET", url, headers=headers, params=querystring)
  stats.append([rounds[idx]])
    
  for i in range(len(response.json()["statistics"][0]['groups'])):
    for j in range(len(response.json()["statistics"][0]['groups'][i]['statisticsItems'])):
      if home[idx] == "home":
        stats[idx].append({response.json()["statistics"][0]['groups'][i]['statisticsItems'][j]['name']:response.json()["statistics"][0]['groups'][i]['statisticsItems'][j]['home']})
      else:
        stats[idx].append({response.json()["statistics"][0]['groups'][i]['statisticsItems'][j]['name'] : response.json()["statistics"][0]['groups'][i]['statisticsItems'][j]['away']})
  

    



```

```python
# https://stackoverflow.com/questions/54943518/create-dataframe-from-list-of-dicts-with-keys-as-index/54943744#54943744

appended = []
for idx , arr in enumerate(stats):
  df = pd.DataFrame.from_dict(data= ChainMap(*stats[idx][1:]), orient='index', columns=[stats[idx][0]])
  appended.append(df)
```

```python
df_concated = pd.concat(appended,axis=1)
```

```python
df_concated
```

```python
df_concated.loc['Results'] = results
```

```python
df_concated
```

```python
df_clean = df_concated.copy()
```

```python
for i in range(6,10):
    df_clean.iloc[i] = (df_concated.iloc[i].str.replace("%","").str.extract(r".*\((.*)\).*").astype('float')/100)[0]
```

```python
df_clean.loc["Ball possession"] = df_concated.loc["Ball possession"].str.replace("%","").astype('float')/100
```

```python
df_clean.fillna(value=0, inplace=True)
```

```python
df_clean
```

```python
df_clean.info()
```

```python
df_clean = df_clean.apply(pd.to_numeric)
```

```python
df_clean.info()
```

```python
df_clean = df_clean.T
```

```python
df_clean.loc[df_clean['Results'] < 0 , 'Results'] = -1
df_clean.loc[df_clean['Results'] == 0, 'Results'] = 0
df_clean.loc[df_clean['Results'] > 0, 'Results'] = 1
```

```python
buttons = [dict(label = 'All',
                  method = 'update',
                  args = [{'visible': True},
                          {'title': 'All',
                           'showlegend':True}])]
for idx , i in enumerate(df_clean.columns):
    lista = [False] *31
    lista[idx] = True
    lista[-1] = True
    buttons.append(
    dict(label = i,
                  method = 'update',
                  args = [{'visible': lista}, 
                          {'title': f'Santos results by {i}',
                           'showlegend':True}, ])
    )
    
    
    
```

```python
layout= go.Layout(
    autosize=False,
    width=1000,
    height=700, title={
        'text': "Santos results by round and game statistics",
        'y':0.9,
        'x':0.5,
        'xanchor': 'right',
        'yanchor': 'top'}  )
    
fig = go.Figure(layout=layout)

for column in df_clean.columns.to_list():
    fig.add_trace(
        go.Scatter(
            x = df_clean.index,
            y = df_clean[column],
            name = column
        )
    )
    
    
    
fig.update_layout(
    updatemenus=[go.layout.Updatemenu(
        active=0,
        buttons=list(
            buttons
            ),
         x = 0.8,
        xanchor = 'left',
        y = 1.1,
        yanchor = 'top'
        )
    ])
fig.show()
```
